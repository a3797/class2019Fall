```python
#parameter setting(variable)
amp=1              #range[0.0,1.0]
sr=10000           #sampling rate, Hz  10000 within 1 second
dur=0.5            #in seconds
freq=100.0         #sine frequency, Hz (similar to male voice), circulation per second
```

```python
t=np.arange(1, sr*dur+1)/sr   #generate time
t
```

```python
theta=t*2*np.pi*freq   #generate phase (time&phase)
```

```python
s=np.sin(theta)  #generate signal by cosine=phasor
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000],'.')
ax.set_xlabel('time(s)')
ax.set_ylabel('real')
```

```python
t=np.arange(1, dur*sr+1)/sr
```

```python
theta=t*2*np.pi*freq
s=np.sin(theta)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000],'.')
ax.set_xlabel('time(s)')
ax.set_ylabel('real')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(t, s,'.')
ax.set_xlabel('time(s)')
ax.set_ylabel('real')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(t[0:100], s[0:100],'.')
ax.set_xlabel('time(s)')
ax.set_ylabel('real')
```

```python
# generate signal by complex-phasor
c = np.exp(theta*1j)   #complex number 1j=i (beyond digit)
c
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t[0:1000], c.real[0:1000], c.imag[0:1000], '.') #real distracts a from a+bi, imaginary distracts b from a+bi
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t[0:100], c.real[0:100], c.imag[0:100], '.') 
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
```
