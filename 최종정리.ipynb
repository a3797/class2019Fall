```python
a=[1,2,3,4]
for i in a:
    print(i)
```

i에는 a의 원소가 하나씩 들어간다. print(a[0]), print(a[1]) 등으로 나누지 않고 한꺼번에 할 수 있도록 도와준다.

```python
for i in range(2):
    print(a[i])
```

```python
for i in range(5):
    print(i)
```

print(i)는 a에 들어가지 않았기때문에 range 설정에 따라서 출력된다. print(a[i])는 range를 a에 할당한다.

```python
for i in range(len(a)):
    print(a[i])
```

```python
a=['red','green', 'blue', 'purple']
b=[0.2, 0.3, 0.1, 0.4]
for i,s in enumerate(a):
    print(s)
```

enumerate는 번호를 매기는 것이다. i에는 index가 들어가고 s에는 a의 값을 할당한다.

```python
for i,s in enumerate(a):
    print(i)
```

```python
for i in enumerate(b):
    print(i)
```

i,s처럼 나누지않을 경우 한꺼번에 출력된다

```python
for i, s in enumerate(a):
    print("{}:{}%". format(s, i*100))
```

```python
for i, s in enumerate(a):
    print("{}:{}%". format(s, b[i]*100))
```

b[i]로 할당하지 않으면 i에는 index값만 할당된다.

```python
for i, s in zip(a,b):
    print("{}:{}%". format(i,s*100))
```

for i,s in zip(a,b)로 쓰면 a의 원소를 i에서 받고 b의 원소를 s에서 받는 루프를 생성한다.

zip에서 a,b로 페어되어야하기때문에 variable은 같아야한다.

```python
for i in range(1,3) :
    for j in range(3,5):
        print(i*j)
```

i는 1,2 j는 3,4이기 떄문에 루프를 돌면 3,4,6,8

```python
for i in range(1,3) :
    print(i)
    for j in range(3,5):
        print(i*j)
```

print(i)로 1을 먼저 출력한 다음 j의 3,4를 곱한 3,4 출력
후에 2를 출력하고 3,4를 다시 곱해서 6,8 출력

```python
for i in range(1,3):
    for j in range(3,5):
        if j >=4 :
            print(i*j)
```

```python
for i in range(1,3):
    if i >=3:
        for j in range (3,5):
            print(i*j)
```

들여쓰기로 묶여있지 않으면 해당되지 않음. >=3에 해당하는 range가 없기 때문에 출력값이 나오지 않는다.
